# .head() --> 
it shows the first N rows in the data ( by default, N=5)
# .shape --> 
it shows the total no. of columns of the dataframe -- not callable
# .index --> 
This Attribute provides the index of the dataframe -- not callable
# .columns --> 
it shows the name of each column -- not callable 
# .dtypes --> 
it shows the data-type of each column -- not callable
# .unique() --> 
In a column, it shows all the unique values, it can be applied on a single column only, not on the whole dataframe
# .nunique() --> 
it shows the total no. of unique values in each column, it can be applied on a single column as well as on a whole dataframe
# .count() -->
it shows the total no. of non-null values in each column, it can be applied on a single column as well as on a whole dataframe
# .value_counts() -->
In a column, it shows all the unique values with their count, it can be applied on a single column only
# .info() -->
Basic info about the dataframe
